```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
FREQUENCY = 100
```

```python
x = np.arange(FREQUENCY * 4)
y = np.zeros_like(x)

y[x % FREQUENCY > (1 - 0.3) * FREQUENCY] = 1

plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title(f"PWM Signal, {FREQUENCY}Hz, 30% duty cycle")
plt.show()
```

```python
x = np.arange(FREQUENCY * 4)
y = np.zeros_like(x)

y[x % FREQUENCY > (1 - 0.6) * FREQUENCY] = 1

plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title(f"PWM Signal, {FREQUENCY}Hz, 60% duty cycle")
plt.show()
```

```python
x = np.arange(FREQUENCY * 4)

z = x.copy()
z = z % FREQUENCY

plt.plot(x, z)
plt.xlabel("Time")
plt.ylabel("Counter")
plt.title("Timer Counter")
plt.show()
```

```python
x = np.arange(FREQUENCY * 4)

y = np.zeros_like(x)
y[x % FREQUENCY > (1 - 0.3) * FREQUENCY] = 1

z = x.copy()
z = (z % FREQUENCY) / FREQUENCY

plt.plot(x, z)
plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Signal and scaled counter")
plt.title("Timer Counter")
plt.show()
```
