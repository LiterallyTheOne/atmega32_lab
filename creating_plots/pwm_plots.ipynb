```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
x = np.arange(50 * 4)
y = np.zeros_like(x)

y[x % 50 > 30] = 1

plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title(r"PWM Signal, 50 Hz, 30% duty cycle")
plt.show()
```

```python
x = np.arange(50 * 4)
y = np.zeros_like(x)

y[x % 50 > 15] = 1

plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title(r"PWM Signal, 50 Hz, 60% duty cycle")
plt.show()
```

```python
x = np.arange(50 * 4)

z = x.copy()
z = z % 50

plt.plot(x, z)
plt.xlabel("Time")
plt.ylabel("Counter")
plt.title("Timer Counter")
plt.show()
```

```python
x = np.arange(50 * 4)

y = np.zeros_like(x)
y[x % 50 > 30] = 1

z = x.copy()
z = (z % 50) / 50

plt.plot(x, z)
plt.plot(x, y)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.title("Timer Counter")
plt.show()
```
