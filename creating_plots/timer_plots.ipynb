```python
from matplotlib import pyplot as plt

import numpy as np
```

```python
x = np.arange(256 * 5)
y = x % 256
l_1 = np.ones_like(x) * 256

plt.figure()

plt.plot(x, y)
plt.plot(x, l_1)

plt.show()
```
